```python
import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. 选择图片和二值图
target = 1
data_name = ['0618', '0854', '1066'][target - 1]
img_path = f'./input_data/{data_name}.png'  # 原图路径
binary_path = f'./input_data/{data_name}_label.png'  # 二值图路径
output_filename = f'./RGB_data/UniformSampler_{data_name}_3x3.csv' #数据输出路径
```

```python
# 加载原图和二值图
img = cv2.imread(img_path)
binary_img = cv2.imread(binary_path, cv2.IMREAD_GRAYSCALE)

# 确定采样步长
step_size = 10  # 增大步长以减少采样点数量

# 数据暂存
features = []
labels = []
sampling_points = []

# 进行均匀采样
height, width, _ = img.shape
for y in range(1, height - 1, step_size):
    for x in range(1, width - 1, step_size):
        # 提取3x3区域的BGR值
        feature_vector = img[y-1:y+2, x-1:x+2].flatten()  # 3x3平铺
        features.append(feature_vector)
        
        # 根据二值图确定标签
        if binary_img[y, x] == 255:  # 假设255代表目标区域
            labels.append(1)  # 标签1代表目标区域
        else:
            labels.append(0)  # 标签0代表背景区域
        
        # 记录采样点位置
        sampling_points.append((x, y))

# 转换为NumPy数组
features_array = np.array(features)
labels_array = np.array(labels)

# 保存数据到CSV文件
df = pd.DataFrame(features_array, columns=[f'R_{i//3 + 1}_{i%3 + 1}' if i % 3 == 0 else f'G_{i//3 + 1}_{i%3 + 1}' if i % 3 == 1 else f'B_{i//3 + 1}_{i%3 + 1}' for i in range(27)])
df['Label'] = labels_array
df.to_csv(output_filename, index=False)

# 画图显示结果
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # 转换BGR为RGB
for (x, y) in sampling_points:
    plt.scatter(x, y, c='yellow', s=50, edgecolors='black')  # 用黑色边缘标记采样点
plt.title('Sampled Points on Original Image')
plt.axis('off')  # 关闭坐标轴
plt.show()

print(f"Data saved to {output_filename}.")
```
