```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def plot_data_points_from_csv(csv_file_path, feature_names):
    # CSV
    data = pd.read_csv(csv_file_path)

    # 20x10
    fig = plt.figure(figsize=(20, 10))

    # 22
    ax1 = fig.add_subplot(221, projection='3d')
    ax2 = fig.add_subplot(222)
    ax3 = fig.add_subplot(223)
    ax4 = fig.add_subplot(224)

    # BGR (3D) 
    # BGR
    ax1.scatter(data[feature_names[0]], data[feature_names[1]], data[feature_names[2]], c=data['Label'], cmap=plt.cm.Paired)
    ax1.set_xlabel(feature_names[0])
    ax1.set_ylabel(feature_names[1])
    ax1.set_zlabel(feature_names[2])
    ax1.set_title('BGR Channels')

    # RG 
    ax2.scatter(data[feature_names[0]], data[feature_names[1]], c=data['Label'], cmap=plt.cm.Paired)
    ax2.set_xlabel(feature_names[0])
    ax2.set_ylabel(feature_names[1])
    ax2.set_title('RG Channels')

    # RB 
    ax3.scatter(data[feature_names[0]], data[feature_names[2]], c=data['Label'], cmap=plt.cm.Paired)
    ax3.set_xlabel(feature_names[0])
    ax3.set_ylabel(feature_names[2])
    ax3.set_title('RB Channels')

    # BG 
    ax4.scatter(data[feature_names[1]], data[feature_names[2]], c=data['Label'], cmap=plt.cm.Paired)
    ax4.set_xlabel(feature_names[1])
    ax4.set_ylabel(feature_names[2])
    ax4.set_title('BG Channels')

    # 
    plt.tight_layout()
    # 
    plt.show()

# csv_file_pathCSV
csv_file_path = './RGB_data/data_0618.csv'  # CSV
plot_data_points_from_csv(csv_file_path, ['B', 'G', 'R'])
```
