```python
import cv2
import numpy as np
from skimage.feature import hog
from skimage import exposure
import matplotlib.pyplot as plt 
from skimage.feature import hog
from skimage import exposure
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report
```

```python
# 数据准备
target = 1
data_name = ['0618', '0854', '1066'][target - 1]

image_path = f'../input_data/{data_name}.png'
label_path = f'../input_data/{data_name}_label.png'
features_output = f'./Features/{data_name}_HOG_featuers.npy'
labels_output = f'./Features/{data_name}_HOG_labels.npy'
output_img_path = f'./img/{data_name}_HOG_img.png'
```

```python
image = cv2.imread(image_path, cv2.IMREAD_COLOR)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
label_image = cv2.imread(label_path, cv2.IMREAD_GRAYSCALE)
```

```python
# 提取HOG特征 (整个图像的HOG特征)
fd, hog_image = hog(gray_image, orientations=9, pixels_per_cell=(8, 8), cells_per_block=(2, 2), visualize=True)

# 进行HOG图像的曝光调整，使得它可视化效果更好
hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10))
# hog_image_rescaled = hog_image

# 可视化HOG特征（可选）
plt.figure(figsize=(10, 5))

# 原始图片
# plt.subplot(1, 2, 1)
# plt.imshow(image)
# plt.title('Original Image')

# HOG特征可视化
# plt.subplot(1, 2, 2)
plt.imshow(hog_image_rescaled, cmap=plt.cm.gray)
# plt.title('HOG Features')
plt.axis('off')
plt.savefig(output_img_path, bbox_inches='tight', pad_inches=0)
plt.show()

# 获取图像的大小
h, w = gray_image.shape

# HOG块的大小
px_per_cell = 8
cells_per_block = 2

# 计算图像中的块的数量
num_cells_x = (w // px_per_cell) - cells_per_block + 1
num_cells_y = (h // px_per_cell) - cells_per_block + 1

# 每个HOG特征块的维度（36维特征）
feature_dim = 36

# 确保HOG特征数与图像块数一致
assert fd.shape[0] == num_cells_x * num_cells_y * feature_dim, f"fd的长度 {fd.shape[0]} 与预期块数 {num_cells_x * num_cells_y * feature_dim} 不匹配！"

# 将fd按每个块的维度分割
fd_reshaped = fd.reshape((num_cells_y * num_cells_x, feature_dim))

# 初始化用于保存特征和标签的列表
features = []
labels = []

# 遍历每个块，提取 HOG 特征并找到对应标签
for i in range(num_cells_y):
    for j in range(num_cells_x):
        # 获取当前块的像素范围
        y_start, y_end = i * px_per_cell, (i + cells_per_block) * px_per_cell
        x_start, x_end = j * px_per_cell, (j + cells_per_block) * px_per_cell

        # 检查块范围是否超出图像边界
        if y_end > h or x_end > w:
            continue

        # 获取当前块对应的掩膜
        block_label_mask = label_image[y_start:y_end, x_start:x_end]
        
        # 获取当前块的主标签（采用多数投票）
        unique, counts = np.unique(block_label_mask, return_counts=True)
        block_label = unique[np.argmax(counts)]  # 选择出现次数最多的标签

        # 获取当前块的 HOG 特征
        block_idx = i * num_cells_x + j
        
        # 获取该块的特征
        block_feature = fd_reshaped[block_idx]

        # 保存特征和标签
        features.append(block_feature)  # 保存整个HOG特征向量
        labels.append(block_label)

# 转换为 NumPy 数组
features = np.array(features)
labels = np.array(labels).reshape(-1, 1)

# 输出 features 和 labels 的形状，确认它们的维度
print(f"features.shape: {features.shape}")
print(f"labels.shape: {labels.shape}")

# 保存到 .npy 文件
np.save(features_output, features)
np.save(labels_output, labels)

print("HOG特征和标签已保存")
```

```python
# 进行 SVM 分类

# 1. 划分数据集
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)

# 2. 标准化特征
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 3. 创建并训练 SVM 分类器
svm = SVC(kernel='linear')  # 使用线性核函数
svm.fit(X_train_scaled, y_train.ravel())  # ravel() 用于将标签展平成一维

# 4. 在测试集上进行预测
y_pred = svm.predict(X_test_scaled)

# 5. 评估模型
print("Accuracy: ", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))

# 6. 在完整图像上应用 SVM 预测
# 这里我们将整个图像分割成块并预测每个块的标签

# 获取预测结果并重塑为图像大小
segmentation_map = svm.predict(scaler.transform(fd_reshaped)).reshape(num_cells_y, num_cells_x)

# 可视化分割结果
plt.imshow(segmentation_map, cmap='gray')
# plt.title('Segmentation Result')
plt.axis('off')
plt.savefig(f'./result/{data_name}_HOG_pred.png', bbox_inches='tight', pad_inches=0.1)
plt.show()
```
