```python
import numpy as np
import cv2
from skimage.feature import graycoprops
from skimage.feature.texture import graycomatrix
from skimage import img_as_ubyte
from skimage.filters import gabor
from skimage.transform import integral_image
import matplotlib.pyplot as plt
```

```python
def compute_gist(image, num_blocks=4, num_orientations=8):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # 1. 将图像分割为多个小块
    h, w = image.shape
    block_h, block_w = h // num_blocks, w // num_blocks
    blocks = []

    for i in range(num_blocks):
        for j in range(num_blocks):
            block = image[i * block_h:(i + 1) * block_h, j * block_w:(j + 1) * block_w]
            blocks.append(block)

    # 2. 使用 Gabor 滤波器提取特征
    # 设置Gabor滤波器的参数
    orientations = np.linspace(0, np.pi, num_orientations, endpoint=False)  # 方向的数量

    gist_features = []
    
    for block in blocks:
        block_features = []
        for angle in orientations:
            real, imag = gabor(block, frequency=0.1, theta=angle)
            block_features.append(np.mean(real))  # 提取实部的均值
        gist_features.append(block_features)
    
    # 3. 将每个块的特征拼接成一个全局特征向量
    gist_vector = np.concatenate(gist_features)
    return gist_vector
```

```python
target = 1
data_name = ['0618', '0854', '1066'][target - 1]
image_path = f'../input_data/{data_name}.png'

# 读取图像，转换为灰度图
image = cv2.imread(image_path)
    
# 计算 GIST 特征
gist_features = compute_gist(image, num_blocks=4, num_orientations=8)
print(gist_features.shape)
print(type(gist_features))
np.save(f'./Features/{data_name}_GIST_featuers.npy', gist_features)

print("GIST 特征向量的长度：", len(gist_features))
    
# 可视化特征
plt.figure(figsize=(10, 5))
plt.bar(range(len(gist_features)), gist_features)
plt.title('GIST Feature Vector')
plt.xlabel('Feature Index')
plt.ylabel('Feature Value')
plt.show()
    
```

```python
import numpy as np
import cv2
from skimage.feature import graycoprops
from skimage.feature.texture import graycomatrix
from skimage import img_as_ubyte
from skimage.filters import gabor
from skimage.transform import integral_image
import matplotlib.pyplot as plt
import json

def compute_gist(image, num_blocks=4, num_orientations=8):
    """
    计算GIST特征
    :param image: 输入图像 (灰度图)
    :param num_blocks: 图像分块数量 (每个维度上块数)
    :param num_orientations: 滤波器的方向数量
    :return: 图像的 GIST 特征向量
    """
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # 1. 将图像分割为多个小块
    h, w = image.shape
    block_h, block_w = h // num_blocks, w // num_blocks
    blocks = []

    for i in range(num_blocks):
        for j in range(num_blocks):
            block = image[i * block_h:(i + 1) * block_h, j * block_w:(j + 1) * block_w]
            blocks.append(block)

    # 2. 使用 Gabor 滤波器提取特征
    # 设置Gabor滤波器的参数
    orientations = np.linspace(0, np.pi, num_orientations, endpoint=False)  # 方向的数量

    gist_features = []
    
    for block in blocks:
        block_features = []
        for angle in orientations:
            real, imag = gabor(block, frequency=0.1, theta=angle)
            block_features.append(np.mean(real))  # 提取实部的均值
        gist_features.append(block_features)
    
    # 3. 将每个块的特征拼接成一个全局特征向量
    gist_vector = np.concatenate(gist_features)
    return gist_vector

# 读取标签图像
def create_label_image(json_path, img_shape):
    """
    从JSON文件创建标签图像
    :param json_path: JSON文件路径
    :param img_shape: 图像的形状 (高, 宽)
    :return: 标签图像 (高, 宽)
    """
    with open(json_path, 'r') as f:
        label_data = json.load(f)
    
    label_image = np.zeros((img_shape[0], img_shape[1]), dtype=np.uint8)
    
    for shape in label_data['shapes']:
        label_name = shape['label']  # 获取标签名称
        points = np.array(shape['points'], dtype=np.int32)  # 获取多边形顶点
        
        # 绘制标签区域
        if label_name == 'road':  # 假设"road"对应标签1
            cv2.fillPoly(label_image, [points], 1)
        elif label_name == 'building':  # 假设"building"对应标签2
            cv2.fillPoly(label_image, [points], 2)
    
    return label_image

# 目标图像选择
target = 1
data_name = ['0618', '0854', '1066'][target - 1]
image_path = f'../input_data/{data_name}.png'
labels_path = f'../input_data/{data_name}.json'

# 读取图像
image = cv2.imread(image_path)
img_height, img_width = image.shape[:2]

# 创建标签图像
labels = create_label_image(labels_path, (img_height, img_width))

# 计算 GIST 特征
gist_features = compute_gist(image, num_blocks=4, num_orientations=8)

# 假设每个块的标签就是该块内的标签（你可以根据需要调整）
# 将标签图像中的每个像素的标签与对应的特征拼接在一起
num_blocks = 4
block_h, block_w = img_height // num_blocks, img_width // num_blocks
block_labels = []

for i in range(num_blocks):
    for j in range(num_blocks):
        # 获取当前块的标签
        block_label = labels[i * block_h:(i + 1) * block_h, j * block_w:(j + 1) * block_w]
        block_labels.append(np.mean(block_label))  # 使用块的平均标签作为该块的标签

# 将特征与标签一一对应
features_with_labels = list(zip(gist_features, block_labels))

# 保存特征和标签
np.save(f'./Features/{data_name}_GIST_features.npy', gist_features)
np.save(f'./Features/{data_name}_GIST_labels.npy', block_labels)

print(gist_features.shape)
block_labels = np.array(block_labels)
print(block_labels.shape)

# 打印信息
print(f"GIST 特征向量的长度：{len(gist_features)}")
print(f"GIST 标签的数量：{len(block_labels)}")

# 可视化特征
plt.figure(figsize=(10, 5))
plt.bar(range(len(gist_features)), gist_features)
plt.title('GIST Feature Vector')
plt.xlabel('Feature Index')
plt.ylabel('Feature Value')
plt.show()
```
