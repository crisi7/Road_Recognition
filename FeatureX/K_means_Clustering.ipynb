```python
import torch 
import numpy as np
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
target = 1
data_name = ['0618', '0854', '1066'][target - 1]
features = np.load(f'./Features/{data_name}_NN_features.npy').squeeze().reshape(512)
print(features.shape)
```

```python
num_clusters = 3
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans = kmeans.fit(features.reshape(-1, 1))

labels = kmeans.labels_
features.shape,labels.shape
```

```python
# 可视化聚类结果（单维特征）
plt.figure(figsize=(8, 6))
plt.scatter(range(len(features)), features, c=labels, cmap="viridis", label="Cluster")
plt.xlabel("Feature Index")
plt.ylabel("Feature Value")
plt.title("K-Means Clustering of Extracted Features")
plt.colorbar(label="Cluster Label")
plt.savefig(f'./KMeans_scatter/{data_name}_NN_KMEANS_SCATTER.png')
plt.show()
```
